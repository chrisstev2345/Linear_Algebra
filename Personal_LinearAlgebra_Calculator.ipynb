```python
import scipy.linalg as la

# Getting the columns
print('\nHow many columns is your matrix? Please enter a positive integer: ')
N = input()
n = int(N)
while n <= 0:
    print('\nThis is not a valid input. Please enter a positive integer.')
    N = input()
    n = int(N)

# Getting the rows
print('\nHow many rows is your matrix? Please enter a positive integer: ')
M = input()
m = int(M)
while m <= 0:
    print('\nThis is not a valid input. Please enter a positive integer.')
    M = input()
    m = int(M)

# Getting the entries
print('\nWhat are the entries for your matrix?')
print('Each column is filled first before proceeding to next column.')
X_0 = []
for j in range(n):
    row = []
    for k in range(m):
        ele = int(input())
        row.append(ele)
    X_0.append(row)
print(X_0)

# Additional information about the matrix entered

if n == m:
    print('\nYou have a square matrix.')
    
    # Find the transpose
    tp_X_0 = np.transpose(X_0)
    print('\nThe transpose is')
    print(tp_X_0)
    
    # Checking if the transpose is equal to original
    if np.allclose(X_0, tp_X_0) == True:
        print('\nYou have a symmetric matrix.')
        print('This means you have real eigenvalues and orthogonal eigenvectors.')
    else:
        print('\nThis is not a symmetric matrix.')
    
    # Calculating the determinant
    det_X_0 = la.det(X_0)
    print('\nThe determinant is')
    print(det_X_0)

    # Finding the inverse
    if det_X_0 == 0:
        print('This matrix is not invertible.')
    else:
        print('\nThe inverse is')
        inv_X_0 = la.inv(X_0)
        print(inv_X_0)
        
        #print('\nChecking if this is the inverse:')
        #calc_iden_1 = np.multiply(X_0, inv_X_0)
        #calc_iden_2 = np.multiply(inv_X_0, X_0)
        #print(calc_iden_1)
        #print(calc_iden_2)
        
    
    # Calculating the trace
    print('\nThe trace of the matrix is')
    print(np.trace(X_0))
    
    # Finding the eigenvalues and eigenvectors
    eig_val, eig_vec = la.eig(X_0)
    print('\nThe eigenvalues are:')
    print(eig_val)
    print('\nThe real values are:')
    print(np.real(eig_val))
    print('\nThe complex values are:')
    print(np.imag(eig_val))
    print('\nThe corresponding eigenvectors are:')
    print(eig_vec)
    
    # Finding the matrix exponential
    from scipy.linalg import expm
    exp_X_0 = la.expm(X_0)
    print('\nThe matrix exponential using a Pade aproximation is:')
    print(exp_X_0)

else:
    tp_X_0 = np.transpose(X_0)
    print('\nThe transpose is')
    print(tp_X_0)



```
